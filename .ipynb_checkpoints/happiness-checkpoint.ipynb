# Transformations

```python
Hapiness_2015_df.columns
```

```python
cols=['Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)','Generosity',]
Hapiness_2015=Hapiness_2015_df[cols].copy()
Hapiness_2015["Year"]=2015
Hapiness_2015.head()
```

```python

Hapiness_2016_df.columns
```

```python
cols=['Country', 'Region', 'Happiness Rank', 'Happiness Score',
       'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity']
Hapiness_2016=Hapiness_2016_df[cols].copy()
Hapiness_2016["Year"]=2016
Hapiness_2016.head()
```

```python
Hapiness_2017_df["Region"]=Hapiness_2016_df["Region"]
Hapiness_2017_df.head()
```

```python
cols=['Country','Region','Happiness.Rank', 'Happiness.Score', 
        'Economy..GDP.per.Capita.', 'Family',
       'Health..Life.Expectancy.', 'Freedom', 'Generosity',
       'Trust..Government.Corruption.']

Hapiness_2017=Hapiness_2017_df[cols].copy()
Hapiness_2017["Year"]=2017


Hapiness_2017.head()
```

```python
Hapiness_2017.rename(columns={"Happiness.Rank":"Happiness Rank","Happiness.Score":"Happiness Score","Economy..GDP.per.Capita.":"Economy (GDP per Capita)",
                            "Health..Life.Expectancy.":"Health (Life Expectancy)","Trust..Government.Corruption.":"Trust (Government Corruption)"},inplace=True)
Hapiness_2017.head()
```

```python
Hapiness_2018_df["Region"]=Hapiness_2015_df["Region"]
Hapiness_2018_df.columns
```

```python
cols=['Country or region','Region', 'Overall rank','Score', 'GDP per capita',
     'Healthy life expectancy','Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
Hapiness_2018=Hapiness_2018_df[cols].copy()
Hapiness_2018.head()
```

```python
Hapiness_2018.rename(columns={"Country or region":"Country","Overall rank":"Happiness Rank","Score":"Happiness Score",
                          "GDP per capita":"Economy (GDP per Capita)","Healthy life expectancy":"Health (Life Expectancy)",
                              "Freedom to make life choices":"Freedom","Perceptions of corruption":"Trust (Government Corruption)"},inplace=True)
Hapiness_2018["Year"]=2018

Hapiness_2018.columns
```

```python
Hapiness_2019_df["Region"]=Hapiness_2015_df["Region"]
Hapiness_2019_df.columns
```

```python
cols=['Country or region','Region','Overall rank', 'Score', 'GDP per capita',
       'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
Hapiness_2019=Hapiness_2019_df[cols].copy()
Hapiness_2019.head()
```

```python
Hapiness_2019.rename(columns={"Country or region":"Country","Overall rank":"Happiness Rank","Score":"Happiness Score",
                             "GDP per capita":"Economy (GDP per Capita)","Healthy life expectancy":"Health (Life Expectancy)",
                             "Freedom to make life choices":"Freedom","Perceptions of corruption":"Trust (Government Corruption)"}, inplace=True)
Hapiness_2019["Year"]=2019

Hapiness_2019.head()
```

```python
Hapiness_2020_df.columns
```

```python
cols=["Country name","Regional indicator","Ladder score","Logged GDP per capita","Healthy life expectancy","Freedom to make life choices","Generosity","Perceptions of corruption"]
Hapiness_2020=Hapiness_2020_df[cols].copy()
Hapiness_2020["Year"]=2020

Hapiness_2020.head()
```

```python
Rank=Hapiness_2020["Ladder score"].rank(ascending=False)
Rank
```

```python
Hapiness_2020['Happiness Rank']=Rank
Hapiness_2020.head(-10)
```

```python
Hapiness_2020.rename(columns={"Country name":"Country","Regional indicator":"Region","Ladder score":"Happiness Score",
                             "Logged GDP per capita":"Economy (GDP per Capita)","Healthy life expectancy":"Health (Life Expectancy)","Freedom to make life choices":"Freedom",
                             "Perceptions of corruption":"Trust (Government Corruption)",}, inplace=True)
Hapiness_2020.head()
```

```python
Happiness_report=pd.DataFrame(columns=["Country","Region","Happiness Rank","Happiness Score","Economy (GDP per Capita)","Health (Life Expectancy)",
                                      "Freedom","Generosity","Trust (Government Corruption)","Year"])
years=[Hapiness_2015,Hapiness_2016,Hapiness_2017,Hapiness_2018,Hapiness_2019,Hapiness_2020]
for i in years:
    Happiness_report=Happiness_report.append(i[["Country","Region","Happiness Rank","Happiness Score","Economy (GDP per Capita)","Health (Life Expectancy)",
                                      "Freedom","Generosity","Trust (Government Corruption)","Year"]],ignore_index='True')
Happiness_report.count()
```

```python
Happiness_report.dropna(inplace=True)
Happiness_report.count()
```

```python
Happiness_report.columns
Happiness_report.head()
```

```python
Happiness_report.rename(columns={"Economy (GDP per Capita)":"GDP"}, inplace=True)
Happiness_report.head()
```

```python
Happiness_report.rename(columns={"Health (Life Expectancy)":"Health","Trust (Government Corruption)":"Trust"},inplace=True)
Happiness_report.head()
```

# Analysis

```python
# Histogram depiction for the Generosity over years
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.5, hspace=1)
figure, axes = plt.subplots(1,5,figsize=(20,5))
plt.subplot(151); plt.hist(x=Hapiness_2015.Generosity, bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2015')
plt.subplot(152); plt.hist(x=Hapiness_2016.Generosity, bins='auto', color='#aaaa00',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2016')
plt.subplot(153); plt.hist(x=Hapiness_2017.Generosity, bins='auto', color='#00e504',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2017')
plt.subplot(154); plt.hist(x=Hapiness_2018.Generosity, bins='auto', color='#a0504a',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2018')
plt.subplot(155); plt.hist(x=Hapiness_2019.Generosity, bins='auto', color='#efc1ff',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2019')
plt.title("Generosity Distribution over years (2015-2019)")
```
